# Only 30 directors are in the data to reduce the complexity

```python
import pandas as pd
import numpy as np



# Step 1: Extract unique director names
if 'director' in df.columns:
    unique_directors = df['director'].unique()
    
    if len(unique_directors) < 30:
        raise ValueError("Not enough unique directors in the column to select 30.")
    
    # Step 2: Randomly select 30 directors
    selected_directors = np.random.choice(unique_directors, size=30, replace=False)

    # Step 3: Fill the director column randomly with the selected directors
    df['director'] = np.random.choice(selected_directors, size=len(df))

    print("Director column filled randomly with 30 unique directors.")
else:
    print("Column 'director' not found in the DataFrame.")

# Display the updated DataFrame
print(df[['director']].value_counts()) 
```

```python
one_hot_df = pd.get_dummies(df['director'], prefix='director', dtype=int)

    # Concatenate the one-hot encoded DataFrame with the original DataFrame
df = pd.concat([df, one_hot_df], axis=1)

    # Step 5: Drop the original director column
df.drop(columns=['director'], inplace=True)

    # Show the updated DataFrame with one-hot encoding
print("\nDataFrame with one-hot encoding:")
print(df)
```

# LET'S FIX GENRES PROBLEM!

```python
unique_genres = df['genres'].unique()
```

```python
lst_genres = []
for genre in list(set(df["genres"].unique())):
    for genre_word in str(genre).split(" "):
        lst_genres.append(genre_word)

```

```python
# Step 2: Clean the list of genres
genres = list(set(lst_genres))
genres.remove("Fiction")
genres[-6] = "Science Fiction"  # Correcting the genre name
print("Genres:", ", ".join(genres))  # Optional, just for checking the genres list

# Step 3: Perform one-hot encoding for each genre
for genre in genres:
    df[genre] = df['genres'].apply(lambda x: 1 if genre in x else 0)

# Print the updated DataFrame
df
```

```python
# Initialize the movie_vector dictionary
movie_vector = {}

# Iterate through each row of the DataFrame
for index, row in df.iterrows():
    movie_title = row['original_title']

    # Values: List of genre indicators (assuming these are at indices 2 to -10)
    genre_values = row[2:-10].tolist()  # Get all genre columns (0s and 1s)
    
    # Initialize a list for vote one-hot encoding
    vote_one_hot = []

    # One-hot encoding for vote averages
    for vote_col in [col for col in row.index if 'vote_average' in col]:
        count = row[vote_col]
        vote_one_hot.append(1 if count > 0 else 0)

    # One-hot encoding for directors (assuming these are the next columns after genres)
    director_values = row[len(genre_values):(len(genre_values) + 30)].tolist()  # Adjust according to your DataFrame structure

    # Combine genre values, one-hot encoding for votes, and director values
    movie_vector[movie_title] = genre_values + vote_one_hot + director_values

# Display the movie vectors
for title, vector in movie_vector.items():
    print(f"{title}: {vector}")
```
